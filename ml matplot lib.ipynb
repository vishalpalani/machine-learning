```python
#matplot import file
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df
```

```python
#scater plot in matplot
plt.scatter(x=df["total_bill"],y=df["tip"])
```

```python
#change colour into red or green
plt.scatter(x=df["total_bill"],y=df["tip"],c="red")
```

```python
#change colour into htmlcolourcode.com to take the colour
#figure is fixing the size of  an cmd
#tittle cmd is use for title
#label cmd is used x lable and ylable
#grid is the grid size to import
plt.figure(figsize=(6,2))
plt.scatter(x=df["total_bill"],y=df["tip"],c="#e1975e",marker="x")
plt.title("billvstip")
plt.xlabel('tip')
plt.ylabel('tbill')
plt.grid()
```

```python
#in mat lab is used seaborn also for figgure also fire import sns and plt
plt.figure(figsize=(6,2))
sns.scatterplot(x=df["total_bill"],y=df["tip"])
```

```python
#bar plot to the program to instead y axis change into height
plt.figure(figsize=(6,2))
plt.bar(x=df["smoker"],height=df["tip"],color="#e1975e")
```

```python
#bar plot ya xis and width used for an particular coloumn
plt.figure(figsize=(6,2))
plt.barh(y=df["smoker"],width=df["tip"],color="#e1975e")
```

```python
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

# make data: correlated + noise
np.random.seed(1)
x = np.random.randn(5000)
y = 1.2 * x + np.random.randn(5000) / 3

# plot:
fig, ax = plt.subplots()

ax.hexbin(x, y, gridsize=20)

ax.set(xlim=(-2, 2), ylim=(-3, 3))

plt.show()
```

```python
x = np.arange(0, 100)
y = x * 2
z = x ** 2
fig, ax = plt.subplots(2, 1)
ax[0].plot(x, y,'r')
ax[0].set_title('Plot of (x, y)')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[1].plot(x, z, 'b')#b is belongs to blue coluour
ax[1].set_title('Plot of (x, z)')
ax[1].set_xlabel('x')
ax[1].set_ylabel('z')


plt.tight_layout()

plt.show()
```
