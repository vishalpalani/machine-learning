```python
# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

# (i) Create a pie chart presenting the male/female proportion
gender_counts = titanic['sex'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)#autopct='%1.1f%%' means that the percentage values displayed on the pie chart will be formatted as floating-point numbers with one decimal place, followed by a percent sign.
plt.title('Male/Female Proportion in Titanic Dataset')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# (ii) Create a scatterplot with the Fare paid and the Age, differ the plot color by gender
plt.figure(figsize=(10, 8))
sns.scatterplot(data=titanic, x='age', y='fare', hue='sex', palette='husl')
plt.title('Scatterplot of Fare paid vs Age, colored by Gender')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Gender')
plt.show()
```

```python
# Load the Diamonds dataset
diamonds = pd.read_csv(r"C:\Users\VISHAL PALANI\Downloads\diamonds.csv")

# Filter diamonds with clarity "SI2" and color "E"
filtered_diamonds = diamonds[(diamonds['clarity'] == 'SI2') & (diamonds['color'] == 'E')]

# Set figure size and font size
plt.figure(figsize=(10, 8))
sns.set(font_scale=1.2)

# Create scatter plot
sns.scatterplot(data=filtered_diamonds, x='carat', y='price', hue='cut', palette='viridis', s=70)

# Set title and labels
plt.title('Relationship between Carat and Price for SI2 Clarity and E Color Diamonds', fontsize=16)
plt.xlabel('Carat (Size of Diamond)', fontsize=14)
plt.ylabel('Price', fontsize=14)

# Customize legend
plt.legend(title='Cut', fontsize=12)

# Show plot
plt.grid(True)
plt.show()
```
